# clean data

```python
ing_domain = ['cuisine_id','salt', 'onions', 'olive oil', 'lime', 'water', 'garlic cloves', 'garlic', 'cachaca', 
 'sugar', 'tomatoes', 'all-purpose flour', 'butter', 'milk', 'eggs', 'unsalted butter', 
 'baking powder', 'large eggs', 'green bell pepper', 'cayenne pepper', 'cajun seasoning', 
 'soy sauce', 'sesame oil', 'corn starch', 'green onions', 'vegetable oil', 'scallions', 'pepper', 
 'oil', 'carrots', 'ground black pepper', 'dried oregano', 'feta cheese crumbles', 'extra-virgin olive oil', 
 'fresh lemon juice', 'garam masala', 'ground turmeric', 'cumin seed', 'ground cumin', 'potatoes', 'baking soda', 
 'grated parmesan cheese', 'ground allspice', 'dried thyme', 'black pepper', 'mirin', 'sake', 'rice vinegar', 
 'ginger', 'sesame seeds', 'chili powder', 'jalapeno chilies', 'sour cream', 'avocado', 'corn tortillas', 'ground cinnamon', 
'ground ginger', 'paprika', 'buttermilk', 'red bell pepper', 'fish sauce', 'coconut milk', 'fresh lime juice', 'shallots']

# ten most common from each cuisine together
```

```python
def clean_ingredient(data, ing_domain):
    cuisines = list(data.cuisine.drop_duplicates())
    
    final_ing = []

    for row in train_list:
        final_ing.append(row[0])
        for ing_interest in ing_domain[1:]:
            final_ing.append(ing_interest in [num for elem in row[1:] for num in elem])

    i=0
    new_list=[]
    while i<len(final_ing):
      new_list.append(final_ing[i:i+64])
      i+=64

    train_tree = pd.DataFrame(new_list, columns = ing_domain, dtype = np.int8)

    return train_tree 
    
```

```python
train_tree = clean_ingredient(train, ing_domain)
```

```python
train_tree.head()
```

# Classification tree; Cuisine ID ~ ingredients

```python
feature_names = list(train_tree.iloc[:,1:].columns)
target_names = list(train_tree.iloc[:,0].drop_duplicates())
```

```python
clf = tree.DecisionTreeClassifier()
clf_fit = clf.fit(train_tree.iloc[:,1:], train_tree.iloc[:,0])
```

```python
type(clf)
```

```python
tree.export_graphviz(clf_fit, out_file='tree.dot')   
```

```python
dot_data = tree.export_graphviz(clf_fit, out_file=None, 
                         feature_names = feature_names,  
                         class_names = target_names,  
                         filled=True, rounded=True,  
                         special_characters=True)  
```
