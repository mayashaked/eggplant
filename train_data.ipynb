```python


import json 
import pandas as pd 
import numpy as np 
from matplotlib import pyplot as plt 
import seaborn as sns
```

```python
train = pd.read_json("C:/Users/river\OneDrive/Desktop/kaggle/train.json")
```

```python
train.sort_values('cuisine')['cuisine'].drop_duplicates().reset_index(drop = True)
```

```python
ing_array = np.array(train['ingredients'])
ingredient_dictionary = {}

for ing_list in ing_array:
    for ing in ing_list:
        if ing in ingredient_dictionary.keys():
            ingredient_dictionary[ing] += 1 
        else:
            ingredient_dictionary[ing] = 1
```
